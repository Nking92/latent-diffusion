```python
from PIL import Image
from base64 import b64decode
import requests
import io
```

```python
result = requests.post('http://localhost:8000/txt2img', json={'prompt': 'a painting of a virus monster playing guitar'}, timeout=600)
data = result.json()
images = data["b64_images"]
for img_b64 in images:
    img_bytes = b64decode(img_b64)
    # img = Image.open(io.BytesIO(img))
    img = Image.open(io.BytesIO(img_bytes))
    img.show()
```
